```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df=pd.read_csv("features.csv")
df.dropna()
```

```python
df['Date'] = pd.to_datetime(df['Date'])

# Visualization 1: Line plot showing the trend of Temperature over time
plt.figure(figsize=(10, 5))
sns.lineplot(x='Date', y='Temperature', data=df, marker='o')
plt.title('Temperature Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Visualization 2: Scatter plot showing the relationship between CPI and Unemployment
plt.figure(figsize=(8, 6))
sns.scatterplot(x='CPI', y='Unemployment', data=df)
plt.title('CPI vs Unemployment')
plt.xlabel('CPI')
plt.ylabel('Unemployment')
plt.grid(True)
plt.show()
```

```python
# Visualization 3: Box plot to visualize the distribution of Fuel Price
plt.figure(figsize=(8, 6))
sns.boxplot(x='Fuel_Price', data=df)
plt.title('Distribution of Fuel Price')
plt.xlabel('Fuel Price')
plt.grid(True)
plt.show()
```

```python
# Visualization 4: Histogram of Markdown1
plt.figure(figsize=(8, 6))
sns.histplot(df['MarkDown1'], kde=True)
plt.title('Histogram of Markdown1')
plt.xlabel('Markdown1')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Visualization 5: Bar plot showing average Markdown2 for each store
average_markdown2 = df.groupby('Store')['MarkDown2'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Store', y='MarkDown2', data=average_markdown2)
plt.title('Average Markdown2 for Each Store')
plt.xlabel('Store')
plt.ylabel('Average Markdown2')
plt.grid(True)
plt.show()
```

```python
# Visualization 6: Pie chart to represent the proportion of holidays
holiday_proportion = df['IsHoliday'].value_counts(normalize=True)
plt.figure(figsize=(6, 6))
plt.pie(holiday_proportion, labels=holiday_proportion.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Holidays')
plt.axis('equal')
plt.show()
```

```python
# Visualization 7: Heatmap to visualize the correlation matrix of the numerical features
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
